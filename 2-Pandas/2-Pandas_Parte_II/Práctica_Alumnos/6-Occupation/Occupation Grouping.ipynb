### Ej 4. Descubre cuál es la edad media por ocupación.

```python
users.groupby('occupation', as_index=False)['age'].mean()
```

### Ej 5. Descubre el ratio de hombres por ocupación y ordénalo de mayor a menor.

```python
num_personas= users.groupby('occupation')['gender'].count() #numero de personas por occupacion
num_hombres= users[users['gender']=='M'].groupby('occupation')['gender'].count() #Numero de hombres por ocupacion

ratio_hombres=num_hombres/num_personas
ratio_hombres.sort_values(ascending=False) #Ordenamos de mayor a menor
```

### Ej 7. Para cada combinación de ocupación y género, calcule la edad media.

```python
users.groupby(['occupation','gender'], as_index=False)['age'].mean().rename(columns={'age':'median_age'})
```

```python
num_personas= users.groupby('occupation')['gender'].count() #numero de personas por occupacion
num_hombres= users[users['gender']=='M'].groupby('occupation')['gender'].count() #Numero de hombres por ocupacion

ratio_hombres=(num_hombres/num_personas)*100 #Porcentaje
ratio_mujeres=100-ratio_hombres

pd.merge(ratio_hombres,ratio_mujeres,on='occupation', suffixes=['_M','_F'])
```

```python
porcent_m = users[users.gender == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
porcent_f = users[users.gender == 'F'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()

porcentajes = pd.DataFrame({'ocupa_m': porcent_m, 'ocupa_f': porcent_f})
porcentajes
```
