```python
import numpy as np
import pandas as pd
```

```python
#Para mostrar multiples dataframes uno al lado del otro
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

### Ej 3. Asignalos a cada una de las siguientes variables: data1, data2, data3.

```python
data1=pd.DataFrame(raw_data_1)
data1
```

```python
data2=pd.DataFrame(raw_data_2)
data2
```

### Ej 4. Une los dos primeros dataframes de datos a lo largo de filas y asígnalos a all_data.

```python
#OPCION 1
all_data= pd.concat([data1,data2], ignore_index=True) #Crea unos indices nuevos para el dataframe resultante
display('data1','data2','all_data')
```

```python
#OPCION 2
all_data= pd.concat([data1,data2]).reset_index(drop=True) #Crea unos indices nuevos para el dataframe resultante
display('data1','data2','all_data')
```

```python
#OPCION 3
pd.merge(data1,data2,how='outer') #Outer: union de todos los datos
```

```python
pd.merge(data1,data2) #Merge NO los concatena. Si no especificamos nada hace la interseccion.
```

```python
#OPCION 1
all_data_col=pd.concat([data1,data2],axis=1) #no tiene mucho sentido hacer esto no?
all_data_col
```

```python
#OPCION 2: hace la union por los subindices de las dos dataframes. como tienen el mismo tamaño esta bien
pd.merge(data1,data2,left_index=True,right_index=True,suffixes=['_data1','_data2'])
```
