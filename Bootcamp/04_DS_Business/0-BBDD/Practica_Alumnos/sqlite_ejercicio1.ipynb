#### Realiza las siguientes consultas

¿Cuáles son los coches más potentes?

```python
#Nombre de los coches mas potentes ordenados de mayor a menos
#Hay signor de interrogacion en algunas variables de horsepower: borrar

comando=''' 
SELECT [car name], CAST(horsepower AS INTEGER) AS horsepower_int
FROM autos
WHERE horsepower != '?'
ORDER BY horsepower_int DESC
'''
pd.read_sql_query(comando, con=conexion)
```

¿Cuántos vehículos tenemos de cada origen?

```python
comando='''
SELECT origin,  COUNT(*) AS cantidad_vehiculos
FROM autos
GROUP BY origin
'''

pd.read_sql_query(comando, con=conexion)
```

¿Cuál es la media de consumo de los vehículos en función de su origen?

```python
comando='''SELECT origin, AVG(mpg) as avg_consumo
FROM autos
GROUP BY origin
'''
pd.read_sql_query(comando, con=conexion)
```

Selecciona los vehículos con un consumo superior a 17 *mpg* y de origen 2.

```python
comando='''
SELECT origin, mpg, [car name]
FROM autos
WHERE origin=2 AND mpg>17
'''
pd.read_sql_query(comando, con=conexion)
```

¿Cuál es la media de peso en aquellos autos con un consumo superior a 15 *mpg*?

```python
comando='''SELECT mpg, AVG(weight) as avg_wight
FROM autos
WHERE  mpg>15
GROUP BY mpg
'''
pd.read_sql_query(comando, con=conexion)
```

La una query de libre elección.

```python
comando='''SELECT cylinders, COUNT(*) AS cantidad_vehiculos
FROM autos
GROUP BY cylinders
'''
pd.read_sql_query(comando, con=conexion)
```
