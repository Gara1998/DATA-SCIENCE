# DECISION TREE



```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

from sklearn import preprocessing, metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn.datasets import load_iris
```

## A) CLASIFICATION

Cargamos los datos y visualizamos el dataset.

```python
iris=load_iris()
#iris

x=iris.data
y=iris.target

df=pd.DataFrame(data=x, columns=iris.feature_names)
df['Target']=y
df.head()



# target_names': array(['setosa', 'versicolor', 'virginica']: 0,1,2
```

Entrenamiento y visualización.

```python
X=df[['petal width (cm)', 'petal length (cm)']]
y=df.Target

tree_clasi=DecisionTreeClassifier(max_depth=2,
                                  random_state=42)

tree_clasi.fit(X,y)
```

```python
X.columns
```

```python
plt.figure(figsize=(8,8))
plot_tree(tree_clasi,
          feature_names=X.columns,
          class_names=list(iris.target_names),
          filled=True);
```

```python
print(tree_clasi.predict_proba([[5,1.9]])) #probabilidad de que pertenezca a cada clase
print(tree_clasi.predict([[5,1.9]])) #A que clase pertenece 
```

## B) REGRESSION

```python
#Conjunto de entrenamiento
np.random.seed(42)
m=200
X=np.random.randn(m,1)
y=4*(X-0.5)**2 + np.random.randn(m,1)/10 #ecuacion cuadratica + ruido

#Arbol de regresión
tree_reg=DecisionTreeRegressor(max_depth=2,
                               random_state=42)

tree_reg.fit(X,y)
```

```python
plt.figure(figsize=(7,7))
plot_tree(tree_reg,
          filled=True);
```
