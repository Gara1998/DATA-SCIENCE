# Gráficos sencillos de líneas

Quizás el más simple de todos los gráficos es la visualización de una función $y = f(x)$.

Aquí veremos por primera vez cómo crear una gráfica simple de este tipo.
Al igual que con todas las secciones siguientes, comenzaremos **configurando el notebook para visualizar e importar los paquetes que usaremos:**

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid') 
import numpy as np
```

Para todos los gráficos de Matplotlib, comenzamos creando una figura y los ejes.
Se pueden crear una figura y ejes de la manera más sencilla de la siguiente manera:

```python
fig = plt.figure()
ax = plt.axes() #de por si asi va de 0 a 1
```

En Matplotlib, la *figura* (una instancia de la clase ``plt.figure``) se puede considerar como un único **contenedor que contiene todos los objetos que representan ejes, gráficos, texto y etiquetas.**
Los *ejes* (una instancia de la clase ``plt.axes``) es lo que vemos arriba: un cuadro delimitador con **marcas y etiquetas, que eventualmente contendrá los elementos de la gráfica que conforman nuestra visualización.**
A lo largo de este notebook, usaremos comúnmente el nombre de variable ``fig`` para referirnos a una instancia de figura y ``ax`` para referirnos a una instancia de ejes o un grupo de instancias de ejes.

Una vez que hayamos creado los ejes, podemos usar la función ``ax.plot`` para trazar algunos datos. Comencemos con una sinusoide simple:

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100) #cien valores del 0 hasta el 10 en la variable x
ax.plot(x, np.sin(x)); # el ; sirve para quitar el texto
```

Como alternativa, podemos usar la interfaz de pylab y dejar que la figura y los ejes se creen en segundo plano.

```python
plt.plot(x, np.sin(x)); #Esto se usa mas para graficas mas sencillas. Pero recomendable usar fig, ax
```

Si queremos crear una sola figura con múltiples líneas, simplemente podemos llamar a la función ``plot`` varias veces:

```python
plt.style.use('tableau-colorblind10') #Usando otro estilo

#Si ejecutamos en una misma celda dos frases se superpondran
plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

¡Eso es todo lo que hay que hacer para visualizar funciones simples en Matplotlib!
Ahora profundizaremos en algunos detalles más sobre cómo controlar la apariencia de los ejes y las líneas.

## Configurando la gráfica: líneas de colores y estilos

La primera configuración que quizás desee realizar en un gráfico es controlar los colores y estilos de las líneas.
La función ``plt.plot()`` toma argumentos adicionales que pueden usarse para especificarlos.
Para ajustar el color, puede utilizar la palabra clave ``color``, que acepta un argumento de cadena que representa prácticamente cualquier color imaginable.
El color se puede especificar de varias maneras:

```python
plt.plot(x, np.sin(x - 0), color='blue')        # Declarando el color por su nombre.
plt.plot(x, np.sin(x - 1), color='g')           # Short color code (rgbcmyk).
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1.
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Sistema hexadecimal (RRGGBB de 00 a FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Escala RGB en formato tupla, valores de 0 a 1.
plt.plot(x, np.sin(x - 5), color='chartreuse'); # Todos los nombres de los colores HTML admitidos.
```

Si no se especifica ningún color, Matplotlib recorrerá automáticamente un conjunto de colores predeterminados para múltiples líneas.

De manera similar, el estilo de línea se puede ajustar usando la palabra clave ``linestyle``:

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# Para abreviar, se pueden utilizar los siguientes códigos:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

Si desea ser extremadamente conciso, estos códigos ``linestyle`` y ``color`` se pueden combinar en un único argumento que no sea una palabra clave para la función ``plt.plot()``:

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

Estos códigos de color de un solo carácter reflejan las abreviaturas estándar en los sistemas de color RGB (rojo/verde/azul) y CMYK (cian/magenta/amarillo/negro), comúnmente utilizados para gráficos digitales en color.

Hay muchos otros argumentos de palabras clave que se pueden utilizar para afinar la apariencia de la gráfica; Para obtener más detalles, sugeriría ver la cadena de documentación de la función ``plt.plot()`` usando las herramientas de ayuda de IPython.

## Configurando la gráfica: límites de los ejes

Matplotlib hace un trabajo decente al elegir los límites de ejes predeterminados para su gráfico, pero a veces es bueno tener un control más preciso.
La forma más básica de ajustar los límites del eje es utilizar los métodos ``plt.xlim()`` (eje x, eje horizontal) y ``plt.ylim()`` (eje y, eje vertical):

```python
x = np.linspace(0,10,100)
x
```

```python
plt.plot(x, np.sin(x))

#De donde a donde queremos que se muestre
plt.xlim(-1.5, 11.5)
plt.ylim(-1.5, 1.5);
```

Si por alguna razón desea que cualquiera de los ejes se muestre al revés, simplemente puede invertir el orden de los argumentos:

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

Un método útil es ``plt.axis()`` (tenga en cuenta aquí la posible confusión entre *axes* con una *e* y *axis* con una *i*).
El método ``plt.axis()`` le permite establecer los límites ``x`` e ``y`` con una sola llamada, pasando una lista que especifica ``[xmin, xmax, ymin, ymax] ``:

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]); #[eje x, eje y]
```

El método ``plt.axis()`` va incluso más allá de esto, permitiéndole hacer cosas como ajustar automáticamente los límites alrededor del gráfico actual:

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

Permite especificaciones de nivel aún más alto, como garantizar una relación de aspecto igual para que en su pantalla, una unidad en ``x`` sea igual a una unidad en ``y``:

```python
plt.figure(figsize=(8,8))
plt.plot(x, np.sin(x))
plt.axis('equal'); # que la longitud del eje x sea igual a la del eje y
```

Para obtener más información sobre los límites de los ejes y otras capacidades del método ``plt.axis``, consulte la cadena de documentación ``plt.axis``.

## Etiquetado de los gráficas

Como última parte de esta sección, veremos brevemente el etiquetado de los gráficos: títulos, etiquetas de ejes y leyendas simples.

Los títulos y las etiquetas de los ejes son las etiquetas más simples; existen métodos que se pueden utilizar para configurarlas rápidamente:

```python
plt.plot(x, np.sin(x))
plt.title("A sin curve")
plt.xlabel("Valores")
plt.ylabel("y");
```

**La posición, el tamaño y el estilo de estas etiquetas se pueden ajustar usando argumentos opcionales para la función.**
Para obtener más información, consulte la documentación de Matplotlib y la documentación de cada una de estas funciones.

Cuando se muestran varias líneas dentro de un solo eje, puede resultar útil crear una **leyenda del trazado de la linea** que etiquete cada tipo de línea.
Nuevamente, **Matplotlib tiene una forma incorporada de crear rápidamente dicha leyenda.**
Se hace mediante el método (lo has adivinado) ``plt.legend()``.
Aunque hay varias formas válidas de usar esto, me resulta más fácil especificar la etiqueta de cada línea usando la palabra clave ``label`` de la función ``plot``:

```python
plt.plot(x, np.sin(x), '-g', label = 'Seno sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'Coseno cos(x)')
plt.axis('equal')
# plt.grid(False)

plt.legend();
```

Como puede ver, la función ``plt.legend()`` realiza un seguimiento del estilo y color de la línea y los relaciona con la etiqueta correcta.
Puede encontrar más información sobre cómo especificar y formatear las leyendas de la gráfica en la documentación de ``plt.legend``.

## Extra: Problemas de Matplotlib

Mientras que la mayoría de las funciones ``plt`` se traducen directamente a métodos ``ax`` (como ``plt.plot()`` → ``ax.plot()``, ``plt.legend()`` → ``ax.legend()``, etc.), este no es el caso para todos los comandos.
En particular, se modifican ligeramente las funciones para establecer límites, etiquetas y títulos.
**Para realizar la transición entre funciones de estilo MATLAB y métodos orientados a objetos, realice los siguientes cambios:**

- ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``

En la interfaz orientada a objetos para visualizar, en lugar de llamar a estas funciones individualmente, suele ser más conveniente usar el método ``ax.set()`` para establecer todas estas propiedades a la vez:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot'); #Sin los dos puntos nos da mas informacion de como se organiza el grafico
```
