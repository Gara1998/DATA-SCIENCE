9. Basándote en el gráfico anterior, escribe encima de cada barra la altura de cada amigo.

plt.text(x,y,valor, va='bottom')

```python
x = ['Ana', 'Luis', 'Pedro', 'Sofía', 'Carmen']
y = [160, 180, 175, 190, 170]
plt.bar(x, y, color='g')

for i in range(5):
    plt.text(x[i], y[i]+8, y[i], va='top', ha='center')
    
plt.xlabel('Amigos')
plt.ylabel('Altura')
plt.title('Altura grupo de amigos');
```

```python
personas={'Ana':160,'Luis':180,'Pedro':175,'Sofia':190,'Carmen':170}
x=personas.keys() #Nombre de personas
y=personas.values()  #Alturas

barplot=plt.bar(x,y,color='green',edgecolor='black')

for dato in barplot: #va barra a barra
    altura=dato.get_height()
    x_b=dato.get_x()
    w_b=dato.get_width()
    plt.text(x_b+w_b/3.5,altura,altura,va='bottom')

plt.bar(x,y,color='green',edgecolor='black')
plt.xlabel('Personas')
plt.ylabel('Alturas')
plt.title('Alturas grupo de amigos');
    
```

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)
nombres = ['Ana', 'Luis', 'Pedro', 'Sofía', 'Carmen']
alturas = [160, 180, 175, 190, 170]

# Crear el gráfico de barras
barrasV=ax.bar(nombres, alturas, color='green', alpha=0.7, edgecolor='black')
#print(barrasV): 5 barras en el container barras
ax.bar_label(barrasV, padding=2,color='black', fontsize=8)

# Agregar etiquetas y título
ax.set_title('Alturas de Amigos en cm',)
```

19. Prueba a cambiar kind a 'pie'

```python
df.plot('Year','Sold_items_A',kind='pie',labels=df['Year'],figsize=(8,8)); #cambiamos tamaño de la figura por que sino se superpone
```

```python
fig, ax = plt.subplots()

maximo=df['Sold_items_A'].max() #QUIERO DESTACAR EL VALOR MAXIMO 7000(%) para destacarlo (lo separo)
elegido=df.loc[df['Sold_items_A']==maximo,'Year'] # CORRESPONDE AL AÑO 2019 que es el maximo

#tengo que conseguir una tupla para pasarla como argumento de la función
lista=[]
for i in range(0,len(df['Sold_items_A'])):  #recorro los valores y al màximo le asigno una separación
    if df['Sold_items_A'][i]==maximo:
        lista.append(0.2) #la separación
    else:
        lista.append(0)   #pegados entre sí
tupla=tuple(lista)  # de [0,0,0,0,0.2] => (0,0,0,0,0.2)

#elegido_destacar = tupla  
# Con una tupla y slider escoger uno a destacar # todos pegados con 0 , la sepacion 0.1,0.2
#autopct='%1.1f%%' con esto da el ehuneko
ax.pie(df['Sold_items_A'],explode=tupla,labels=df['Year'],autopct='%1.3f%%',shadow=True, startangle=180); #explode=separacion de las 
#pocentaje el ángulo(se ve distinto)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))
data = [1000, 3500, 4000, 5500, 7000]
ingredients = ['2015', '2016', '2017', '2018', '2019']
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}".format(absolute)  # "{:.1f}%\n({:d})"
wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), labels=ingredients)
ax.legend(wedges, ingredients,
          title="Ingredients",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=15, weight="bold")
ax.set_title("Matplotlib bakery: A pie")
plt.show()
```

```python
plt.pie(df['Sold_items_A'], labels=df['Year'], autopct=lambda p: '{:.0f}\n({:.1f}%)'.format(p * sum(df['Sold_items_A']) / 100, p), startangle=90)
plt.axis('equal')  # Para asegurarse de que el gráfico sea circular
plt.title("Distribución de Ventas en el Departamento A por Año")
plt.show()
```
