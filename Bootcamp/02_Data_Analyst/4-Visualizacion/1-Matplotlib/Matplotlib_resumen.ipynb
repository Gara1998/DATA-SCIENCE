## HISTOGRAMAS

La mayoria de los valores de la distribuci√≥n normal estandar se encuentran entre el -3 y el 3. Con una n al final de randN creamos una distribucion normal.

```python
data=np.random.randn(1000) 
plt.hist(data);
```

```python
plt.hist(data, bins=200, alpha=0.8, color='red');
```

```python
# distribuci√≥n normal con una media de 0 y una desviaci√≥n est√°ndar de 0.8. 
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(alpha = 0.5, bins = 20)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

Si deseamos contar el numero de puntos en una seccion del histograma pero sin que nos de el dibujo:
- np.histogram: Esta funci√≥n calcula un histograma a partir de un conjunto de datos. Toma dos argumentos principales: el conjunto de datos (data en este caso) y el n√∫mero de bins en los que se dividir√° el rango total de los datos (bins=5 en este caso).
- counts: los conteos de cu√°ntos elementos de data caen en cada uno de los 5 bins del histograma.
- bins_egges: bordes de los bins

```python
counts, bins_edges=np.histogram(data, bins=5)
print(counts)
print('-'*100)
print(bins_edges)
```

### Histogramas bidimensionales

- np.random.multivariate_normal: Genera n√∫meros aleatorios siguiendo una distribuci√≥n multivariada normal. Toma tres argumentos principales: la media (mean), la matriz de covarianza (cov), y el n√∫mero de muestras a generar (10000 en este caso).

- .T: Transpone los resultados, es decir, intercambia filas por columnas. Esto es √∫til para desempaquetar los resultados en dos arreglos x e y.

```python
mean = [0, 0] #media
cov = [[1, 1], [1, 2]] #matriz de covarianza
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist(x);
#plt.hist(y)
```

```python
plt.hist2d(x, y, bins = 30, cmap = 'Blues');
cb = plt.colorbar()
cb.set_label('counts in bin');
```

Con secciones hexagonales:

```python
plt.hexbin(x, y, gridsize = 10, cmap = 'Blues')
cb = plt.colorbar(label='count in bin');
```

Analizar el histograma sin dibujarlo:

```python
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
```

```python
display(xedges,yedges,counts)
```

```python
counts_x,edges_x=np.histogram(x,bins=10)
display(edges_x,counts_x)
counts_y,edges_y=np.histogram(y,bins=10)
display(edges_y,counts_y)
```
