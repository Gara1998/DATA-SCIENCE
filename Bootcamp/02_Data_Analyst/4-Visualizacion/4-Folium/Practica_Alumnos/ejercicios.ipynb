#### 3. Añade una marca sobre las estaciones de bicicletas respetando las sigueintes indicaciones.
* Las marcas tienen que respetar la siguiente relación de colores:
    * vacío --> rojo
    * lleno --> azul
    * otro --> verde
* Deben llevar el icono de bicicletas.
* Al hacer click debe aparecer el nombre de la estación y el número de bicicletas.

```python
latitud=df['Latitude'].mean()
longitud=df['Longitude'].mean()

#Nos ubicamos en las coordenadas que queramos
mapa=folium.Map(location=[latitud,longitud], zoom_start=14, control_scale=True)

#Añadimos marcas sobre las estaciones de bicicletas
for (index,row) in df.iterrows():
    if row.loc['Bikes_Available']==0:
        folium.Marker(location=[row.loc['Latitude'], row.loc['Longitude']],
                      icon=folium.Icon(color='red', icon='bicycle ', prefix='fa'),
                      popup=row.loc['Name']+', '+str(row.loc['Bikes_Available'])).add_to(mapa)
    elif row.loc['Bikes_Available']==30:
        folium.Marker(location=[row.loc['Latitude'],row.loc['Longitude']],
                      icon=folium.Icon(color='blue', icon='bicycle ', prefix='fa'),
                      popup=row.loc['Name']+', '+str(row.loc['Bikes_Available'])).add_to(mapa)
    else:
        folium.Marker(location=[row.loc['Latitude'],row.loc['Longitude']],
                      icon=folium.Icon(color='green', icon='bicycle ', prefix='fa'),
                      popup=row.loc['Name']+', '+str(row.loc['Bikes_Available'])).add_to(mapa)
mapa
```
